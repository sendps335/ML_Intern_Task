## Importing the Required Libraries

```
import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
import seaborn as sns
```

## Loading the Dataset

```
df=pd.read_csv("/content/country_vaccinations.csv")
df.head(10)
```

As We Can See that the the datasets consists of the amount of vaccinations provided to people in various countries on a daily basis.
So it is a time-series dataset where the values can be grouped by the countries

```
df['country'].describe()
```

# Visualization

```
country=set(df['country'])
for i in country:
  df_data=df[df['country']==i]
  df_data['total_vaccinations'].fillna('NaN',inplace=True) #Unnecessary but essential step
  avg=0
  vac=list(df_data['total_vaccinations'])
  # Moving Average
  for j in range(1,len(vac)-1):
    if(vac[j] == 'NaN'):
      vac[j]=avg
    else:
      avg=(avg*j + vac[j])/(j+1)
  print(vac)
  plt.plot(df_data['date'],vac,color='cyan')
  plt.title(i)
  plt.xticks(df_data['date'],"")
  plt.xlabel("Time->")
  plt.ylabel('Total Vaccinations')
  plt.show()
  print()
```

## Lets See the distributions of various vaccines in different countries


```
country=set(df['country'])
for i in country:
  df_data=df[df['country']==i]
  df_data['people_vaccinated'].fillna('NaN',inplace=True) #Unnecessary but essential step
  avg=0
  vac=list(df_data['people_vaccinated'])
  # Moving Average
  for j in range(0,len(vac)):
    if(vac[j] == 'NaN'):
      vac[j]=avg
    else:
      avg=(avg*j + vac[j])/(j+1)
  print(vac)
  plt.plot(df_data['date'],vac,color='cyan')
  plt.title(i)
  plt.xticks(df_data['date'],"")
  plt.xlabel("Time->")
  plt.ylabel('people_vaccinated')
  plt.show()
  print()
```
